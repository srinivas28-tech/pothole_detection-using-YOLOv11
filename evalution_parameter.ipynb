```python
from ultralytics import YOLO

# 5️⃣ Load the trained model
model = YOLO("best.pt")  # Since 'best.pt' is inside 'Yolov11'

# 6️⃣ Validate the model on the test dataset
metrics = model.val(data="data.yaml")  # Update if data.yaml has a different name

# Extract accuracy metrics
print(f"mAP@50: {metrics.box.map50:.4f}")  # Mean Average Precision at 50% IoU
print(f"mAP@50-95: {metrics.box.map:.4f}")  # Mean Average Precision (50-95%)

# Convert numpy arrays to scalars
box_precision = metrics.box.p.mean()  # Mean precision for all classes
box_recall = metrics.box.r.mean()  # Mean recall for all classes

seg_precision = metrics.seg.p.mean()  # Mean precision for segmentation
seg_recall = metrics.seg.r.mean()  # Mean recall for segmentation

# Print final results
print(f"Box Precision: {box_precision:.4f}")  # Precision for object detection
print(f"Box Recall: {box_recall:.4f}")  # Recall for object detection
print(f"Mask Precision: {seg_precision:.4f}")
print(f"Mask Recall: {seg_recall:.4f}")
```
